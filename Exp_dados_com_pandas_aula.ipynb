```python
colunas_traduzidas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=colunas_traduzidas)
print("Colunas do DataFrame traduzidas:")
print(df.columns)
```

```python
mapeamento_senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Diretor'
}
df['senioridade'] = df['senioridade'].map(mapeamento_senioridade)
print("Contagem de valores na coluna 'senioridade' após a tradução:")
print(df['senioridade'].value_counts())
```

```python
mapeamento_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato Temporário',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].map(mapeamento_contrato)
print("Contagem de valores na coluna 'contrato' após a tradução:")
print(df['contrato'].value_counts())
```

```python
mapeamento_tamanho_empresa = {
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
}
df['tamanho_empresa'] = df['tamanho_empresa'].map(mapeamento_tamanho_empresa)
print("Contagem de valores na coluna 'tamanho_empresa' após a tradução:")
print(df['tamanho_empresa'].value_counts())
```

```python
mapeamento_remoto = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
}
df['remoto'] = df['remoto'].map(mapeamento_remoto)
print("Contagem de valores na coluna 'remoto' após a tradução:")
print(df['remoto'].value_counts())
```

```python
import numpy as np

# criação de base de teste
df_salarios = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "salario": [4000, np.nan, 5000, np.nan, 100000]
})
# essa linha calcula a media e substitui os valores NaN por medias  e aredonda o valor
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))
# Calcula a mediana e substituia a=os NaN pelas medianas
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())
df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})
# repete o dia anterior para os dias nulos
df_temperaturas["preenchido_ffill"] = df_temperaturas["temperatura"].fillna(method="ffill")
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})
# Repete o dia posterior para o dias nulos
df_temperaturas["preenchido_bfill"] = df_temperaturas["temperatura"].fillna(method="bfill")
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    "nome": ["Ana", "Bruno", "Carlos", "Daniele", "Val"],
    "cidade": ["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]
})
# Essa linha altera e troca o valor nulo por um valor que escolhemos ex: Não Informado
df_cidades["cidade_preenchida"] = df_cidades["cidade"].fillna("Não Informado")
df_cidades
```
